### From the above file `goog.csv`: Let's Return the Date, too

Lets get back to the function we did on computing the highest value in a column. 
A tuple type is useful here if we want to return the date in addition to the highest value! 

We index the values the same way we do with lists, 0 for the first item and 1 for the second, etc.

Just like we did before, first we have to reload our data into the mydata variable

```python
def read_data(filename):
    # Function takes file path, and returns a list of the rows of the data
    data = []
    with open(filename, errors="ignore") as file:
        for row in file:
            data.append(row)
    return data

mydata = read_data("data/goog.csv")
```

Now instead of initializing highest to 0, we initialize it to the tuple (0,0). The rest is the same, except for `highest = (vals[0], float(vals[column]))` where we also save the date of the new highest value.

```python
def get_highest_with_date(data, column):
#print("my column number is", column)
# take a list of rows and a column number, then return a tuple pair
    highest = (0,0)
    for row in data[1:]:   # why do we start with 1 and not 0?  Check the file for a tip...
        vals = row.split(",")
        # remember these are the cols: 'Date,Open,High,Low,Close,Volume\n'
        if float(vals[column]) > highest[1]:
            #print("my row is now", row)
            #print(highest)
            # we check the value in the second element using highest[1].
            # then we assign the tuple to highest, using parens.  vals[0] is the date!
            highest = (vals[0], float(vals[column]))
            #print("after assignment:", highest)
    # now we return the tuple
    return highest
```

If we want to use the results from the function, we must save to a variable or variables:

```python
date, value = get_highest_with_date(mydata, 1)
print("The date of the highest value was {}".format(date)) 
print("The highest value was {}".format(value))
```

You can also save them into a single variable, but the resulting variable will still be a tuple, because we created it when we did this: `highest = (vals[0], float(vals[column]))`.  Those parentheses mean tuple with 2 items!

### Storing data in a dictionary

```python
def make_dict(data):
    mydict = {}
    for row in data:
        vals = row.split(',')
        mydict[vals[0]] = vals[1:]# mydict(key) = values
    return mydict
```

Notice there is a bad "\n" character in here. Also, notice that some of the lists don't have all the same number of elements.  Volume is missing from a bunch of dates. Let's use "strip" to clean the "\n" up and also do a bunch of list comprehensions to clean and change the data to numbers:

```python
mydata =mydata[1:]
```

```python
def make_dict(data):
    mydict = {}
    for row in data:
        vals = row.split(',')
        # this is a list comprehension that applies strip to remove whitespace chars including
        # \n from each value in the vals list, then reassigns the result to vals, overwriting
        # the previous messy list.
        vals = [val.strip() for val in vals]
        # another list comprehension -- now we turn everything into a floating point number,
        # except for the first value, which is the date string and except for the empty strings.
        # There is an if-test to rule out the '' empty strings!
        floats = [float(val) for val in vals[1:] if val != '']
        # now we save the floats list to the dictionary using the date, the first item, as key
        mydict[vals[0]] = floats
    return mydict
```

```python
dictdata = make_dict(mydata)
```

### Ordered dictionary

Remember that dictionaries aren't ordered. But our data is inherently ordered by date.  Luckily, there is a special type of dictionary called OrderedDict that will retain order for us.  

An OrderedDict is a dictionary subclass that remembers the order in which its contents are added.

A regular dict does not track the insertion order, and iterating over it produces the values in an arbitrary order. 

In an OrderedDict, by contrast, the order the items are inserted is remembered and used when creating an iterator.

It lives in the collections library, just like the useful Counter does.  The documentation is here: https://docs.python.org/3/library/collections.html#collections.OrderedDict




```python
from collections import OrderedDict
```

We create an ordered dictionary of our dataset using a sorted version of the dict we made, sorting by the first value in items, 
which is the date:

```python
ordereddata = OrderedDict(sorted(dictdata.items(), key=lambda x: x[0])) #sort according to date x[0]
```
